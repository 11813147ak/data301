## Data visualization

Visualization itself is a very broad area, with *data visualization* being only a very narrow area with a particular goal. The goal of *data visualization* is to create a visual representation of data to enable humans to extract understanding and answer substantive analytical questions about the data. Data visualizations are not mere pictures to look at. Their goal is to help, and even provoke, us think about complex topics. While a data visualization may, at times, be entertaining or aesthetically pleasing, those are not its primary goals. Our introduction to data visualization begins with a description of *visualization grammar* or the *grammar of graphics*.

According to the [Cambridge Dictionary](http://dictionary.cambridge.org/us/dictionary/english/grammar), *grammar* is:

> The study or use of the rules about how words change their form and combine with other words to express meaning.

A visualization grammar is then:

> The study or use of the rules about how the components of a visualization combine to express aspects of the underlying dataset.

Here are some good starting points for reading about visualization grammars:

- [Polaris](https://graphics.stanford.edu/papers/polaris_extended/polaris.pdf), Chris Stolte, Diance Tang and Pat Hanrahan (2002).
- [The Grammar of Graphics](https://www.amazon.com/Grammar-Graphics-Statistics-Computing/dp/0387245448), Leland Wilkinson (2005).
- [A Layered Grammar of Graphics](http://vita.had.co.nz/papers/layered-grammar.pdf), Hadley Wickham (2010).

For a quick introductory read, Hadley Wickham's short paper is probably best.

## Grammar

The visualization grammar we will work with here is based on the abstractions in the visualization libraries, [d3](https://d3js.org/) and [Vega](https://vega.github.io/). This grammar has the following components:

![](images/viz_grammar.png)

- **Data:** The input data to visualize
- **Transforms:** Grouping, agg, stats, projection, layout
- **Scales:** Map data values to visual properties
- **Guides:** Axes and legends visualize scales
- **Marks:** Visual elements to represent data

To better understand visualization grammar, we show how these components are manifest in some well known and effective visualizations.

## Example: WSJ's Measles

http://graphics.wsj.com/infectious-diseases-and-vaccines/
