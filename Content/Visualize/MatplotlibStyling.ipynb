# Styling Matplotlib Visualizations

**Learning Objectives:** Learn how to style Matplotlib plots using Matplotlib style sheets and Seaborn.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Styling using Matplotlib's style sheets

If you have version 1.4.0 or greater of Matplotlib, you can style plots using Matplotlib's `style` package.

```python
import matplotlib
```

```python
matplotlib.__version__
```

To see the available styles, look at the `style.available` attribute:

```python
plt.style.available
```

To use of of these styles, simply call `style.use`:

```python
plt.style.use('ggplot')
```

```python
x = np.linspace(0.0, 3.141, 100)
noise = np.random.randn(100)
```

```python
plt.plot(x, 0.5*x + 0.1*noise)
```

```python
plt.style.use('bmh')
```

```python
plt.plot(x, 0.5*x + 0.1*noise)
```

You can also create and share your own Matplotlib style sheets. For further information see the [Style Sheet Documentation](http://matplotlib.org/users/style_sheets.html).

## Styling using Seaborn

Seaborn is a visualization library that is built on top of Matplotlib but adds:

* Advanced statistical plotting.
* Beatiful styling.

In this notebook, you will learn how to style standard Matplotlib plots using Seaborn. The statistical plotting features of Seaborn will be covered elsewhere.

To enable Seaborn's styling, simply import it:

```python
import seaborn as sns
```

And then use Matplotlib's standard plotting functions:

```python
plt.plot(x, 0.5*x + 0.1*noise)
```

The change the Seaborn styling using the `set_style` function with any of the following options:

* `darkgrid`
* `whitegrid`
* `dark`
* `white`
* `ticks`

To change the size of the figures use the `set_context` function with any of the following options:

* `paper`
* `notebook`
* `talk`
* `poster`

To reset everything use `sns.set()`.

```python
sns.set_style('darkgrid')
sns.set_context('talk')
```

```python
plt.plot(x, 0.5*x + 0.1*noise)
```

Here is an example of a histogram using the default Seaborn styling:

```python
sns.set()
```

```python
plt.hist(np.random.randn(1000), bins=50);
```

You can use the `despine` function *for the current plot* to remove the axis spines:

```python
sns.set_style('ticks')
```

```python
plt.plot(2.0*x, np.sin(2.0*x))
plt.axis([-0,7,-1.5,1.5])
sns.despine(offset=10)
```
