# Plotting Distributions of Data

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

**Learning Objectives**: Learn to visualize distributions of numerical data using histograms, box plots, violin plots and scatter plots and facet/group those plots by categorical variables.

## Load data sets

For this notebook, we will use the tips data set from Seaborn:

```python
tips = sns.load_dataset('tips')
tips['tip_pct'] = tips.tip/tips.total_bill
```

```python
tips.head()
```

## Single one variable numerical distribution: `hist` and `distplot`

The simplest type of distribution is a single numerical variable. This corresponds to a one dimensinal probability distribution. The most common way of visualizing such a distribution is a histogram. You can always create a histogram using Matplotlib's `hist` function:

```python
plt.hist(tips.tip)
plt.xlabel('Tip ($)')
plt.title('Tip Amount');
```

For more advanced distribution plotting, Seaborn has a `distplot` function:

```python
sns.distplot?
```

Here is a default distribution plot using `distplot`. Notice how it shows the histogram along with a smooth curve that approximates the distribution. This smooth sure is called a [kernel density estimation](http://en.wikipedia.org/wiki/Kernel_density_estimation) or KDE. The `distplot` also uses a more advanced method than `hist` for pcking the default number of bins.

```python
sns.distplot(tips.tip);
```

The `distplot` function can show the distribution using three different approaches:

* Histogram (`hist`)
* Rug plot (`rug`)
* Kernel density (`kde`)

Here is a KDE+rug version of the distribution:

```python
sns.distplot(tips.tip, rug=True, kde=True, hist=False);
plt.title('Distribution of Fares');
```

## Comparing numerical distributions: `boxplot` and `violinplot`

```python
sns.boxplot?
```

With a single argument that is a `DataFrame`, the different columns of the `DataFrame` are used for the different boxes. If you use this approach, you will usually want to have two columns with similar data. In the tips data set, we don't really have that, but we can look at the `total_bill` and `tip` columns anyways:

```python
sns.boxplot(tips[['total_bill','tip']]);
```

Much of the time, you will want to create boxes using a `groupby` operation. This is done by passing a second argument to `boxplot` that will be used for grouping.

Here we are looking at the `fare` distribution, grouped by `sex`:

```python
sns.boxplot(tips.tip, tips.sex);
```

By passing a list of columns as the second argument, you can group by multiple variables:

```python
sns.boxplot(tips.tip_pct, [tips.sex, tips.day], vert=False);
```

The `violinplot` function has a similar API as `boxplot`, but shows the distributions using "violins," which are mirrored and rotated histograms. Here is the documentation:

```python
sns.violinplot?
```

```python
sns.violinplot(tips.tip_pct, [tips.sex, tips.smoker]);
```

## Two numerical distributions: `scatter` and `jointplot`

If you want to plot a two-dimensional distribution (2 numerical variables), you can always use Matplotlib's `scatter` function:

```python
plt.scatter(tips.total_bill, tips.tip)
plt.xlabel('total_bill')
plt.ylabel('tip');
```

However, Seaborn has a `jointplot` function that has a number of nice enhancements, such as showing the marginal distributions.

```python
sns.jointplot?
```

```python
sns.jointplot('total_bill','tip',data=tips);
```

The `jointplot` function has a `kind` argument that can be used to control how the distributions are visualized. Here is a KDE version of the distributions, using the `white` style:

```python
with sns.axes_style('white'):
    sns.jointplot('total_bill', 'tip', data=tips, kind='kde');
```

And a hex plot version:

```python
sns.jointplot('total_bill', 'tip', data=tips, kind='hex');
```

## Facetting distribution plots

Where Seaborn really shines is in its ability to quickly created plots that are facetted or grouped by different categorical variables. The main interface for doing this is the `FacetGrid` function. Here is the full [FacetGrid Documentation](http://web.stanford.edu/~mwaskom/software/seaborn/tutorial/axis_grids.html#subsetting-data-with-facetgrid).

```python
sns.FacetGrid?
```

Here is the general approach for using `FacetGrid`:

1. Create a `FacetGrid` object:
  - Pass your `DataFrame` as the first argument
  - Use the `row`, `col` and `hue` arguments to choose the variables you will facet/group by.
2. Use the `.map()` method to call the particular plotting function
  - The first argument is the plotting function to call to make each plot
  - The subsequent arguments are the arguments to that function
  - Use column names to specify what column to use

### Facetted histograms

Here is a simple histogram of the `tip_pct` variable, grouped by the `sex` column:

```python
fg = sns.FacetGrid(tips, col='sex', size=4)
fg.map(plt.hist, 'tip_pct');
```

In this case, we will group by two variables (`sex`, `smoker`) and use a KDE based `distplot`. Notice how we have used the `margin_titles` and `size` arguments to tune the visual style.

```python
fg = sns.FacetGrid(tips, row='sex', col='smoker', margin_titles=True, size=4)
fg.map(sns.distplot, 'tip_pct', rug=False, hist=False, kde=True)
fg.add_legend()
```

### Facetted scatter plots

The `FacetGrid` function also works with scatter plots. This offers and extremely efficient API for visualizing numerical distributions.

```python
fg = sns.FacetGrid(tips, row='sex', col='smoker', hue='day', size=4, margin_titles=True)
fg.map(plt.scatter, 'total_bill', 'tip', alpha=0.8)
fg.add_legend()
```
